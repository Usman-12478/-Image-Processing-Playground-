# Rotated Image

```python
from scipy.ndimage import rotate
plt.figure(figsize = (6, 4))
rotated_image = rotate(grayscale_image, angle = 45, mode = 'reflect')
plt.imshow(rotated_image)
plt.axis('off')
plt.show()
```

```python
rotated_image1 = rotate(grayscale_image, angle = 90, mode = 'reflect')
plt.imshow(rotated_image1)
plt.axis('off')
plt.show()
```

```python
heigt_scale = 200 / grayscale_image.shape[0]
width_scale = 200 / grayscale_image.shape[1]

resized_image = zoom(grayscale_image, (heigt_scale, width_scale))
plt.figure(figsize = (6, 4))
plt.imshow(resized_image, cmap = 'gray')
plt.axis('off')
plt.show()
```

# Blurred Image

```python
blurred_image = gaussian_filter(grayscale_image, sigma = 3)
plt.figure(figsize = (6, 4))
plt.imshow(blurred_image, cmap = 'gray')
plt.axis('off')
plt.show()
```

```python
blurred_image1 = gaussian_filter(grayscale_image, sigma = 2)
plt.figure(figsize = (6, 4))
plt.imshow(blurred_image1)
plt.axis('off')
plt.show()
```

```python
sobel_1 = sobel(grayscale_image, axis = 1)
sobel_2 = sobel(grayscale_image, axis = 0)

edge_image = np.hypot(sobel_1, sobel_2)
plt.figure(figsize = (6, 4))
plt.imshow(edge_image, cmap = 'gray')
plt.axis('off')
plt.show()
```
